```python
import numpy as np

def Rocchio(alpha, beta, gamma):
    
    d0 = "king palace jungle sleeps"
    d1 = "jungle lion fire king discovery"
    d2 = "king timon musafa"
    d3 = "timon lion simba"
    
    q = "lion king"
    
    terms = d0.split() + d1.split() + d2.split() + d3.split() + q.split()
    terms = sorted(set(terms))

    term_doc_matrix = np.zeros((len(terms), 5))
    documents = [d0, d1, d2, d3, q]
        
    for i, doc in enumerate(documents):
        for j, term in enumerate(terms):
            term_doc_matrix[j, i] = doc.split().count(term)
    
    document_vectors = term_doc_matrix[:, :4]
    query_vector = term_doc_matrix[:, 4]
    
    relevant_docucments_idx = [1]
    non_relevant_documents_idx = [3]
    
    relevant = np.mean(document_vectors[:, relevant_docucments_idx], axis=1)
    non_relevant = np.mean(document_vectors[:, non_relevant_documents_idx], axis=1)

    print(term_doc_matrix)
    print("\n")
    print(query_vector)
    print("\n")
    print(relevant)
    print("\n")
    print(non_relevant)
    print("\n")
    
    new_query_vector = (alpha * query_vector) + (beta * relevant) - (gamma * non_relevant)
    
    return terms, new_query_vector
```

```python
alpha = 1.0
beta = 0.8
gamma = 0.1

terms, new_query_vector = Rocchio(alpha, beta, gamma)
print("Terms", terms)
print("Updated Q Vector", new_query_vector)
```
