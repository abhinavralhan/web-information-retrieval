```python
import numpy as np
```

```python
def suggest_words(partial_input, word_list):
    suggested_words = []
    for word in word_list:
        if word.startswith(partial_input):
            suggested_words.append(word)
    return suggested_words
```

```python
def autosuggest():
    file_path = input("Enter the file path of the list with the suggestion words: ")

    try:
        with open(file_path, 'r') as file:
            word_list = file.read().splitlines()
    except FileNotFoundError:
        print("File not found.")
        return

    while True:
        partial_input = input("Enter a partial word or phrase: ")

        suggested_words = suggest_words(partial_input, word_list)  

        if suggested_words:
            print("Suggestions:")
            for word in suggested_words:
                print(word)
        else:
            print("No suggestions found.")

        choice = input("Enter 's' to select a suggestion, or 'c' to continue typing: ")

        if choice == "s" and suggested_words:
            selected_word = input("Enter the number of the suggestion you want to select: ")
            if selected_word.isdigit() and int(selected_word) in range(1, len(suggested_words) + 1):
                print("Selected suggestion:", suggested_words[int(selected_word) - 1])
            else:
                print("Invalid selection.")
        elif choice == "c":
            continue
        else:
            print("Invalid choice. Exiting the program.")
            break

        print()  # Print an empty line for better readability

# Run the autosuggest program
autosuggest()
```
