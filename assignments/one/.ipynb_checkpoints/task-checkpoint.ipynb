```python
survived = data[data['survived'] == 'yes']
survived.gender.value_counts()
```

```python
died = data[data['survived'] == 'no']
died.gender.value_counts()
```

```python
data['class_size'] = data['class']
survived_by_class = data.groupby(['class']).agg({'class_size': 'count'}).reset_index().merge(data.groupby(['class']).survived.agg(lambda x: (x=='yes').sum()).reset_index().merge(data.groupby(['class']).survived.agg(lambda x: (x=='no').sum()).reset_index(), how='left', on='class', suffixes=('_yes', '_no')), how='left', on='class')
survived_by_class['survived_ratio'] = survived_by_class['survived_yes'] / survived_by_class['class_size']
survived_by_class['died_ratio'] = survived_by_class['survived_no'] / survived_by_class['class_size']
survived_by_class
```
