```python
def precision_func(matrix):
    retrieved = [x[0] for x in matrix]
    gold_standard = [x[1] for x in matrix]
    
    tp_a = 0
    tp_b = 0
    tp_c = 0
    
    fp_a = 0
    fp_b = 0
    fp_c = 0
    
    for prediction, label in zip(retrieved, gold_standard):
        if prediction == label:
            if prediction == 'setosa':
                tp_a += 1
            elif prediction == 'versicolor':
                tp_b += 1
            elif prediction == 'virginica':
                tp_c += 1
        else:
            if prediction == 'setosa':
                fp_a += 1
            elif prediction == 'versicolor':
                fp_b += 1
            elif prediction == 'virginica':
                fp_c += 1

    precision_setosa = tp_a / (tp_a + fp_a)
    precision_versicolor = tp_b / (tp_b + fp_b)
    precision_virginica = tp_c / (tp_c + fp_c)
    
    print("Precision for Setosa: ", precision_setosa)
    print("Precision for Versicolor: ", precision_versicolor)
    print("Precision for Virginica: ", precision_virginica)
    
    return precision_setosa, precision_versicolor, precision_virginica
```

```python
def recall_func(matrix):
    retrieved = [x[0] for x in matrix]
    gold_standard = [x[1] for x in matrix]
    
    tp_a = 0
    tp_b = 0
    tp_c = 0
    
    fn_a = 0
    fn_b = 0
    fn_c = 0
    
    for prediction, label in zip(retrieved, gold_standard):
        if prediction == label:
            if prediction == 'setosa':
                tp_a += 1
            elif prediction == 'versicolor':
                tp_b += 1
            elif prediction == 'virginica':
                tp_c += 1
        else:
            if prediction == 'setosa':
                if label == 'versicolor':
                    fn_b += 1
                else:
                    fn_c += 1
            elif prediction == 'versicolor':
                if label == 'setosa':
                    fn_a += 1
                else:
                    fn_c += 1
            elif prediction == 'virginica':
                if label == 'setosa':
                    fn_a += 1
                else:
                    fn_b += 1

    recall_setosa = tp_a / (tp_a + fn_a)
    recall_versicolor = tp_b / (tp_b + fn_b)
    recall_virginica = tp_c / (tp_c + fn_c)

    print("Recall for Setosa: ", recall_setosa)
    print("Recall for Versicolor: ", recall_versicolor)
    print("Recall for Virginica: ", recall_virginica)
    return recall_setosa, recall_versicolor, recall_virginica
```
