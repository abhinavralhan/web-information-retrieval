```python
import pandas as pd
import numpy as np
import matplotlib
from sklearn.datasets import load_iris
```

```python
d1 = ["python", "anaconda", "jupyter", "programming", "learn", "computer"]

d2 = ["game", "are", "world", "programming", "learn", "learn", "are"]

d3 = ["learn", "are", "world", "programming"]
```

```python
def my_TD_matrix(d1, d2, d3):
    combined_lists = d1 + d2 + d3
    terms = set(combined_lists)
    
    term_document_matrix = {}
    for term in terms:
        term_matrix = []
        document1_term_count = [1 if term in d1 else 0]
        document2_term_count = [1 if term in d2 else 0]
        document3_term_count = [1 if term in d3 else 0]
        
        term_matrix = document1_term_count + document2_term_count + document3_term_count
        term_document_matrix[term] = term_matrix
            
    return term_document_matrix
```

```python
term_document_matrix = my_TD_matrix(d1, d2, d3)
term_document_matrix
```

```python
term_document_matrix['python'], term_document_matrix['learn']
```

```python
python_td_matrix = term_document_matrix['python']
learn_td_matrix = term_document_matrix['learn']

query_result = [python_td_matrix[i] and learn_td_matrix[i] for i in range(len(python_td_matrix))]
print(query_result)
```
